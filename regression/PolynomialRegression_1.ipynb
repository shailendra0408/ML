```python
import pandas as pd
import warnings
import numpy as np
#for Visualization 
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_excel(r'/Users/shailendra0408/Desktop/CSA_Temp.xlsx')
print (df.shape)
```

```python
RH = df['Relative Humidity']
SR = df['Solar Radiation']
T = df['Temperature']
X = df[["Relative Humidity", 'Solar Radiation']]
y = df['Temperature']
```

```python
#Visualizing the data 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
axes[0].scatter(RH,y)
axes[1].scatter(SR,y)
axes[0].set_title("RH Plotted")
axes[1].set_title("SR Plotted")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3,random_state=42)


```

```python
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_train, y_train)

```

```python
poly_reg_y_predicted = poly_reg_model.predict(X_test)
poly_reg_rmse = np.sqrt(mean_squared_error(y_test, poly_reg_y_predicted))
print (poly_reg_rmse )
print (poly_reg_model.coef_)
print (X_test)
print(poly_reg_y_predicted)
```
